```python
import pandas as pd
import numpy as np
#Task 1
np.random.seed(42)
data = {
'video_id': [f'V{i:03d}' for i in range(1, 21)],
'views_start': np.random.randint(1000, 10000, 20),
'views_end': np.random.randint(3000, 50000, 20),
'category': np.random.choice(['Юмор', 'IT', 'Еда'], 20)
}
df = pd.DataFrame(data)
df.to_csv('tiktok_videos.csv', index=False)

print(df.head())
```

```python
# 1. Рассчитать growth_rate — процент роста просмотров за день
df['growth_rate'] = ((df['views_end'] - df['views_start']) / df['views_start']) * 100

# 2. Отфильтровать видео с ростом более 200%
hype_df = df[df['growth_rate'] > 200]
hype_df.head()
```

```python
# 3. Подсчитать количество хайповых видео по категориям
hype_counts = hype_df['category'].value_counts().reset_index()
hype_counts.columns = ['Категория', 'Количество']

# 4. Вывести ТОП-3 самых быстрорастущих видео
top_3 = df.nlargest(3, 'growth_rate')

print("--- Хайповых видео по категориям ---")
print(hype_counts.to_string(index=False))
print("\n--- ТОП-3 самых быстрорастущих видео ---")
print(top_3[['video_id', 'category', 'growth_rate']].to_string(index=False))
```

```python
#Task 3

messages = [
    "Привет всем участникам чата",
    "Розыгрыш iPhone прямо сейчас!!!",
    "Кто знает хороший курс по Python?",
    "@admin помогите разобраться",
    "КРИПТА только растет #bitcoin $$$",
    "Отличная статья про машинное обучение",
    "Розыгрыш* призов каждый день",
    "Подскажите книгу для начинающих",
    "Заработок на крипте легко",
    "Спасибо за помощь",
    "### Супер предложение ###",
    "Изучаю pandas для анализа данных",
    "Розыгрыш 1000$ для всех",
    "Какой фреймворк выбрать для веба?",
    "Крипта это будущее финансов",
]
df_chat = pd.DataFrame({'message': messages})
```

```python
df_chat['is_spam'] = df_chat['message'].str.contains(r'крипт|розыгрыш', case=False, na=False)
df_chat['message_clean'] = df_chat['message'].str.replace(r'[@#$*]', '', regex=True)
df_chat['word_count'] = df_chat['message_clean'].str.split().str.len()
df_filtered = df_chat[(df_chat['word_count'] > 3) & (~df_chat['is_spam'])]
df_filtered[['message_clean', 'word_count']]
```

```python
#Task 5
np.random.seed(42)
users = [f'User_{i}' for i in range(1, 11)]
genres = ['Рок', 'Поп', 'Техно', 'Рэп']
data = np.random.randint(0, 100, size=(10, 4))
df_music = pd.DataFrame(data, columns=genres, index=users)
```

```python
music_matrix = df_music.to_numpy()
row_sums = music_matrix.sum(axis=1, keepdims=True)
norm_matrix = music_matrix / row_sums
user_1_profile = norm_matrix[0]
distances = np.sqrt(np.sum((norm_matrix - user_1_profile)**2, axis=1))
distances[0] = np.inf
closest_idx = np.argmin(distances)
best_match = df_music.index[closest_idx]

print(f"Наиболее похожий пользователь: {best_match} (расстояние: {distances[closest_idx]:.4f})")
print("\nСравнение профилей (нормированных):")
print(f"User_1:   {norm_matrix[0]}")
print(f"{best_match}:   {norm_matrix[closest_idx]}")
```
